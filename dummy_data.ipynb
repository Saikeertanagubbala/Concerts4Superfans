```python
import requests
import re
import pandas as pd
from bs4 import BeautifulSoup

response = requests.get("https://www.billboard.com/charts/hot-100/", 
                       headers={"User-Agent": "Mozilla/5.0"})
soup = BeautifulSoup(response.text, "html.parser")
artists = [a.text.strip() for a in soup.select(".c-label.a-no-trucate")][:100]
artists = [re.split(r'&|[Ww]ith|[Ff]eaturing', a)[0].strip() for a in artists]
artists = list(dict.fromkeys(artists))[:50]
```

```python
df = pd.DataFrame(artists, columns =["artist_name"])
df.to_csv("billboard_top_artists_sample_data.csv", index=False)
```
